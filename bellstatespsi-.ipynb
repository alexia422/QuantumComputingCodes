```python
from qiskit import *
from qiskit.visualization import plot_histogram
```

```python
qr = QuantumRegister(2)
cr = ClassicalRegister(2)
circuit4 = QuantumCircuit(qr,cr)
circuit4.x(1)
circuit4.h(0)
circuit4.z(0)
circuit4.z(1)
circuit4.cx(0,1)
circuit4.measure(qr,cr)
circuit4.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
results4 = execute(circuit4, backend= simulator).result()
```

```python
plot_histogram(results4.get_counts(circuit4))
```
